```python
#!/usr/bin/env python

import sympy as sp

def solve_4_site_symbolically():
    # Define symbols
    E     = sp.Symbol('E', complex=True)       # eigenvalue
    gamma = sp.Symbol('gamma', real=True)      # non-Hermitian parameter
    t1    = sp.Symbol('t1', real=True, positive=True)
    t2    = sp.Symbol('t2', real=True, positive=True)
    
    # Define the 4x4 Hamiltonian H
    #  H = [
    #    [ -i gamma,   t1,         0,      0    ],
    #    [    t1,      -i gamma,   t2,     0    ],
    #    [     0,         t2,     -i gamma, t1 ],
    #    [     0,         0,        t1,     0   ]
    #  ]
    # In Sympy, we write -i gamma as -I*gamma:
    H = sp.Matrix([
        [-sp.I*gamma,   t1,         0,         0        ],
        [      t1,  -sp.I*gamma,    t2,        0        ],
        [       0,       t2,    -sp.I*gamma,   t1       ],
        [       0,       0,         t1,        0        ]
    ])

    # Characteristic polynomial p(E) = det(H - E I)
    pE = (H - sp.eye(4)*E).det()

    # Condition for an EP: p(E) = 0 and d/dE p(E) = 0
    dp_dE = pE.diff(E)

    # Solve the system of equations: p(E) = 0, dp_dE = 0
    # for E and gamma, in terms of t1, t2.
    # This can get complicated, so we might specify t1, t2 numerically,
    # or see if Sympy can handle the full symbolic solution.
    sol = sp.solve([pE, dp_dE], [E, gamma], dict=True)

    return sol

if __name__ == "__main__":
    solutions = solve_4_site_symbolically()

    print("Symbolic solutions for the 4-site model's EP conditions:")
    for s in solutions:
        print(s)
```

```python
#!/usr/bin/env python

import numpy as np
import matplotlib.pyplot as plt

def H_4_site(t1, t2, gamma, E=0):
    """
    Construct the 4x4 Hamiltonian for given t1, t2, gamma.
    """
    # Define complex i as 1j in Python
    mat = np.array([
        [-1j*gamma,  t1,        0,         0     ],
        [ t1,    -1j*gamma,     t2,        0     ],
        [ 0,       t2,     -1j*gamma,     t1     ],
        [ 0,        0,         t1,        0      ]
    ], dtype=complex)
    return mat

def find_ep_numerically(t1, t2, gamma_min=0.0, gamma_max=5.0, steps=1000):
    """
    Sweep gamma from gamma_min to gamma_max in 'steps' increments,
    diagonalize the Hamiltonian, and look for level collisions.
    Returns an array of (gamma, min_gap) where min_gap is the
    minimum separation between any two eigenvalues at that gamma.
    """
    gammas = np.linspace(gamma_min, gamma_max, steps)
    results = []
    for g in gammas:
        H = H_4_site(t1, t2, g)
        eigvals = np.linalg.eigvals(H)
        # Compare all pairwise differences
        # We'll measure "closeness" in the complex plane (absolute difference).
        min_dist = np.inf
        for i in range(len(eigvals)):
            for j in range(i+1, len(eigvals)):
                dist = np.abs(eigvals[i] - eigvals[j])
                if dist < min_dist:
                    min_dist = dist
        results.append((g, min_dist))
    return np.array(results)

if __name__ == "__main__":
    t1_val = 1.0
    t2_val = 0.5

    data = find_ep_numerically(t1_val, t2_val, gamma_min=0.0, gamma_max=3.0, steps=500)
    gammas = data[:,0]
    min_gaps = data[:,1]

    # Find the gamma that minimizes the gap
    idx_min = np.argmin(min_gaps)
    gamma_at_ep = gammas[idx_min]

    print("Approx. EP found near gamma =", gamma_at_ep, "with gap =", min_gaps[idx_min])

    # Optionally plot the minimal gap vs gamma
    plt.plot(gammas, min_gaps, label="Min gap between any two eigenvalues")
    plt.axvline(gamma_at_ep, color="r", linestyle="--", label=f"EP approx at gamma={gamma_at_ep:.2f}")
    plt.xlabel("gamma")
    plt.ylabel("Minimum pairwise eigenvalue separation")
    plt.title("Searching for EP (level coalescence) in 4-site model")
    plt.legend()
    plt.show()
```
